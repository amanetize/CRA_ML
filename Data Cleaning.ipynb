```python
import pandas as pd

# Load the Excel file
xls_file = pd.ExcelFile('/content/drive/MyDrive/CRA/credit-risk.xlsx')

# Read the relevant sheets, ignoring 'Description' and 'Sheet8'
data = pd.read_excel(xls_file, sheet_name=[name for name in xls_file.sheet_names
                                           if name not in ('Description', 'Sheet8')])

# Create an empty list to hold the DataFrames
df_list = []

# Append data from the relevant sheets into the list
for n_rows in ('1-20k', '20k-40k', '40k-60k', '60k-80k', '80k-100k', '100k-121k'):
    df_list.append(data[n_rows])

# Concatenate all DataFrames into one
companies_df = pd.concat(df_list, ignore_index=True)

companies_df
```

```python
original_len = int(len(companies_df))
companies_df = companies_df.dropna()
print('Removed rows:', str(original_len - int(len(companies_df))))
```
